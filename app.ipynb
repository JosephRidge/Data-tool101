```python
# Source - https://stackoverflow.com/a/73617021
# Posted by jezrael, modified by community. See post 'Timeline' for change history
# Retrieved 2026-07-03, License - CC BY-SA 4.0

#https://stackoverflow.com/a/71515035/2901002
#  count total stock itemsm

all_stock_df = legit_df["StockCode"].value_counts() 
def autopct_format(values):
        def my_format(pct):
            total = sum(all_stock_df)  
            val = int(round(pct*total/100.0))
            return '{:.1f}%\n({v:d})'.format(pct, v=val)
        return my_format

s = legit_df["StockCode"].value_counts().head()
plt.pie(s,labels = s.index, autopct=autopct_format(s))
#pandas only alternative
#s.plot.pie(autopct=autopct_format(s))
```

```python
legit_df["Country"].value_counts().head()
```

```python
all_Country_df = legit_df["Country"].value_counts() 
def autopct_format(values):
        def my_format(pct):
            total = sum(all_Country_df)  
            val = int(round(pct*total/100.0))
            return '{:.1f}%\n({v:d})'.format(pct, v=val)
        return my_format

s = legit_df["Country"].value_counts().head(3)
plt.pie(s,labels = s.index, autopct=autopct_format(s))
```
